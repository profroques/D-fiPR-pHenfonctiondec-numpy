<h1>Tracer l'évolution du pH en fonction de c</h1>

<h2>Premier script demandé</h2>
<h3>Consignes</h3>
<p>
<b>Saisir</b> un script permettant :

1. de <b>créer</b> une liste appelée "c" contenant 4 valeurs de concentrations en quantité de solutions filles : 0,1 ; 0,01 ; 0,001 et 0,0001 mol/L<br>
2. de <b>créer</b> une liste appelée "pHexp" contenant les valeurs des pH mesurés des solutions filles précédentes : 0,9 ; 2,1 ; 2,95 et 4,07<br>
3. d'<b>afficher</b> les valeurs de la liste c<br>
4. d'<b>afficher</b> les valeurs de la liste pHexp<br>
5. de <b>créer</b> une liste appelée "pHtheo" qui correspond au pH des solutions à partir du calcul pH = - log (c) / le calcul est réalisé par l'intermédiaire de la fonction `log10` de la bibliothèque **math**<br>
6. d'<b>afficher</b> les valeurs de la liste "pHtheo"<br>
7. de <b>créer</b> une liste, à partir d’une bouce `for` et  appelée “cth” qui contiendrait un ensemble de valeur de concentration comprise entre 0,0009 mol/L et 0,15 mol/L par pas de 10-4 mol/L
8. de <b>créer</b> une liste, à partir de la même boucle que la liste “cth” et appelée “pHth” qui contiendrait les valeurs des pH (= - log (cth)) de chaque valeur de cth
9. d'<b>afficher</b> sur un même graphique : pHtheo en fonction de c (type courbe rouge), pHexp en fonction de c (nuage de points "+" bleu) et pHth en fonction de cth (type courbe verte)
---
<h3><span style="color:red">A propos de la fonction log10 de la bibliothèque <b>math</b></span></h3>

* Ne pas oublier d'importer la bibliothèque math. Ex : `from math import *`
* La fonction `log10(x)` retourne le logarithme en base 10 de x


```python
from math import*
import matplotlib.pyplot as plt

c = [0.1,0.01,0.001,0.0001]
pHexp = [0.9,2.1,2.95,4.07] 
print('valeurs de c : ', c)
print()
print("Valeurs du pH : ", pHexp)
print()
pHtheo=[]
for i in c :
    pHtheo.append(-log10(i))

print("valeurs du pH calculé à partir de - log c : ", pHtheo)
print()

cth=[] #création d'une liste pour des concentrations comprises entre 0,00009 et 0,15 mol/l
pHth = []
for i in range((9),15000,10):
    cth.append(i/100000)
    pHth.append(-log10(i/100000))

#Tracé de l'évolution du pH en fonction de la concentration
plt.plot(c,pHtheo,color = 'red', label = 'modelisation théorique à partir des 4 valeurs') #Tracé du modèle théorique
plt.plot(cth,pHth,color = 'green', label = 'modelisation théorique') #Tracé du modèle théorique
plt.scatter(x=c,y=pHexp,marker= '+', label='Mesures expérimentales') #Tracé des points correspondants aux mesures expérimentales
plt.xlabel('Concentration en mol/L')
plt.ylabel('pH')
plt.title('Evolution du pH en fonction de la concentration')
plt.legend()
```

<h3><span style="color:red">A propos de la fonction log10 de la bibliothèque <b>numpy</b></span></h3>

* Ne pas oublier d'importer la bibliothèque math. Ex : `import numpy as np *`
* La fonction `log10(x)` retourne le logarithme en base 10 de x - en cas d'usage d'un alias du type np pour la bibliothèque numpy, il faut alors saisir : np.log10(x)


```python
import numpy as np
import matplotlib.pyplot as plt

c = [0.1,0.01,0.001,0.0001] 
pHexp = [0.9,2.1,2.95,4.07] 
print('valeurs de c :', c)
print()
print("Valeurs du pH mesuré : ", pHexp)
print()
pHtheo=[]
for i in c :
    pHtheo.append(-np.log10(i))

print("valeurs du pH calculé à partir de - log c : ",pHtheo)
print()

cth = np.arange(0.00009,0.15,0.0001) #Génère un ensemble de valeurs de concentrations
pHth=-np.log10(cth)

#Tracé de l'évolution du pH en fonction de la concentration
plt.plot(c,pHtheo,color = 'red', label = 'modelisation théorique à partir des 4 valeurs de c') #Tracé du modèle théorique
plt.plot(cth,pHth,color = 'green', label = 'modelisation théorique') #Tracé du modèle théorique
plt.scatter(x=c,y=pHexp,marker= '+', label='Mesures expérimentales') #Tracé des points correspondants aux mesures expérimentales
plt.xlabel('Concentration en mol/L')
plt.ylabel('pH')
plt.title('Evolution du pH en fonction de la concentration')
plt.legend()
```
